# Assignment : Plot different graphs/charts with the data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = 'https://raw.githubusercontent.com/hakimyameen/Datasets/main/fifa_eda.csv'
df = pd.read_csv(path)
df.head()
```

```python
df.info()
```

```python
df.describe().T
```

## Who are the oldest players still playing?

```python
data1 = df[['Name','Age']].nlargest(10,'Age')
plt.figure(figsize=(15,4))
plt.bar(data1.Name,data1.Age)
plt.xlabel('Players')
plt.ylabel('Age')
plt.title('Oldest Players')
plt.show();
```

## Which positions have the maximum players?

```python
data2 = df['Position'].value_counts()
plt.figure(figsize=(10,7))
plt.xlabel('Number of Players')
plt.ylabel('Positions')
plt.title('Total Number of Players for different Positions')
plt.barh(data2.index,data2.values);
```

## A look at the height distribution of the players

```python
plt.hist(df['Height'])
plt.xlabel('Height')
plt.ylabel('No. of Players')
plt.title('Height distribution of Players')
plt.grid();
```

## Comparison of International Reputation Vs Wages

```python
plt.scatter(df['International Reputation'],df.Wage)
plt.xticks(df['International Reputation'].unique(),df['International Reputation'].unique().astype(str))
plt.xlabel('International Reputation')
plt.ylabel('Wages')
plt.title('International Reputation Vs Wages');
```

## Which nation produces the maximum number of football players?

```python
data3 = df['Nationality'].value_counts()[:10]
plt.pie(data3,labels=data3.index,autopct='%.2f%%');
```

## Countries with the tallest players:

```python
data4 = df.groupby('Nationality').mean()[['Height']].nlargest(10,'Height')
data4
plt.figure(figsize=(15,5))
plt.plot(data4)
plt.xlabel('Country')
plt.ylabel('Avg height of players')
plt.title('Countries with the tallest players')
plt.grid();
```
