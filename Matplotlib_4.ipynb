# Assignment: Create a dataframe with 5 rows & 4 columns using random.random
# and do pie plot
# and do check transpose of the same

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
d = pd.DataFrame(np.random.random((5,4)), columns=['data1','data2','data3','data4'])
d
```

## Pie Plot

```python
d.plot(kind='pie', subplots=True, layout=(3,2), figsize=(10,10), autopct='%.f%%');
```

## Pie Plot of Transpose

```python
d.T.plot(kind='pie', subplots=True, layout=(3,2), figsize=(10,10), autopct='%.f%%');
```
